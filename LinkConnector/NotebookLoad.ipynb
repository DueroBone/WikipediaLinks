```python
import networkx as nx

# import plotly.graph_objects as go
import pickle
import gzip
import os
import time
```

```python
print("Starting loading")
start_time = time.time()
G: nx.DiGraph = pickle.load(open("wikipedia_graph.pkl", "rb"))
print(
    f"Graph loaded from wikipedia_graph.pkl in {time.time() - start_time:.2f} seconds"
)
```

```python
betweenness = nx.betweenness_centrality(G)

print("Sorting...")
# Sort nodes by betweenness (descending)
sorted_nodes = sorted(betweenness.items(), key=lambda x: x[1], reverse=True)

# Choose top n nodes
n = 5
top_n_nodes = sorted_nodes[:n]
```

```python
print(top_n_nodes)
```

```python
with open("betweenessGraph.pkl", "wb") as f:
    pickle.dump(betweenness, f)
```
